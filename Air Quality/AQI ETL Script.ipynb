```python
import urllib3
import requests
import certifi
import json
import datetime
import pandas as pd
import authentication as a
http = urllib3.PoolManager()
```

```python
start_date = datetime.date(2020, 10, 1)
number_of_days = 182
datelist = [(start_date + datetime.timedelta(days = day)).isoformat() for day in range(number_of_days)]
header_tf = True
columns_to_drop = ['HourObserved', 'LocalTimeZone', 'Category.Number','Category.Name'
                   ,'ReportingArea','StateCode', 'Latitude', 'Longitude']
column_order = ['DateObserved','MonthObserved','DayObserved','ParameterName','AQI']
url_b_date = 'https://www.airnowapi.org/aq/observation/zipCode/historical/?format=application/json&zipCode=84120&date='
url_a_date = 'T00-0000&distance=25&API_KEY=' + a.api_auth
for date in datelist:
    if date != datelist[0]:
        header_tf = False
    url = url_b_date + date + url_a_date
    r = http.request('GET', url)
    # decode json data into a dict object
    data = json.loads(r.data.decode('utf-8'))
    df = pd.json_normalize(data)
    df = df.drop(columns_to_drop, axis=1)
    df['DateObserved'] = pd.to_datetime(df['DateObserved'])
    df['MonthObserved'] = df['DateObserved'].dt.to_period('M')
    df['DayObserved'] = pd.DatetimeIndex(df['DateObserved']).day
    df = df[column_order]
    df.to_csv (r'aqi_stuffs.csv', mode='a', index = False, header=header_tf)
```
